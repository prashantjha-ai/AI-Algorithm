# 3. Findings

## Number of matches played by each team

```python
data["date"] = pd.to_datetime(data["date"])

# Creating new column year the dates
data["Year"] = data["date"].dt.year
# Creating new column month the dates
data["Month"] = data["date"].dt.month

# Creating new column day the dates
data["Day"] = data["date"].dt.day

# Creating new column weekday the dates
data["Weekday"] = data["date"].dt.weekday
```

```python
# Creating the variable list as matches and converting that into Series
matches = list(data["team1"]) + list(data["team2"])
matches = pd.Series(matches)

matches_df = pd.DataFrame(matches)

# Number of matches played by each team
matches.value_counts().sort_values(ascending = False)

plt.figure(figsize=(15,4))
sns.histplot(matches)
plt.xticks(rotation = 89)
plt.xlabel("Teams", fontsize = 14)
plt.ylabel("Number of matches", fontsize = 14)
plt.title("Number of matches played by each team", fontsize = 14)
plt.show()
```

## Numbers of matches won by each team

```python
data.winner.value_counts()

sns.barplot( x = data.winner.value_counts(), y = data.winner.value_counts().index)
plt.xlabel("Number of wins", fontsize = 14)
plt.ylabel("Teams",fontsize = 14)
plt.title("Matches won by each team", fontsize = 14)
plt.show()
```

## Team's winning percentage for each Year

```python
for i in data.Year.unique():
  print("*********  Year :", i,"  *********")
  plt.pie(data[data["Year"]== i]["winner"].value_counts(), autopct = "%1.1f%%")
  plt.legend(data[data["Year"]== i]["winner"].value_counts().index, bbox_to_anchor = (1,1),title='Teams in {}'.format(i) ) # The bbox_to_anchor argument is used to specify the precise placement of the legend.
  plt.show()
```
